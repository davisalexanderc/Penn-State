```python
import matplotlib.pyplot as plt
import pandas as pd
import ast

def plot_histogram_for_image(filename, csv_path):
    """
    Plot histograms for rod and sphere areas for a given image file.

    Args:
        filename (str): The name of the image file to look up.
        csv_path (str): The path to the CSV file containing the data.

    """
    # Load the CSV file
    path = pd.read_csv(csv_path)

    # Find the row corresponding to the specified filename
    row = path[path['filename'] == filename]

    if row.empty:
        print(f"Filename {filename} not found in the CSV file.")
        return

    # Extract rod and sphere areas from the row
    rows_rod = row.iloc[0]['rod_area']
    rows_sphere = row.iloc[0]['sphere_area']

    # Convert string representation of lists into actual lists
    data1 = ast.literal_eval(rows_rod)
    data2 = ast.literal_eval(rows_sphere)

    # Plot both histograms on top of each other
    if data1:
        plt.hist(data1, bins=20, alpha=0.5, label='Rod Area', color='blue', edgecolor='black')
    if data2:
        plt.hist(data2, bins=20, alpha=0.5, label='Sphere Area', color='red', edgecolor='black')
    plt.title('Histogram of Nanoparticle Areas')
    plt.xlabel('Area')
    plt.ylabel('Frequency')
    plt.legend(loc='upper right')
    plt.show()

# Example usage
filename = 'CN-017_000003.jpg'
csv_path = './outputs/output.csv'
plot_histogram_for_image(filename, csv_path)
```
